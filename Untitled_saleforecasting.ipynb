```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



df=pd.read_csv('C:\\Users\\ms\\PycharmProjects\\pythonProject1\\/advertising.csv')




print("Data shape:", df.shape)
s
sns.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', kind='scatter')
plt.title('Scatterplot Matrix')
plt.xlabel('Advertising Budget (TV, Radio, Newspaper)')
plt.ylabel('Sales')
plt.show()


df['TV'].plot.hist(bins=10, alpha=0.5, label='TV')
df['Radio'].plot.hist(bins=10, color="green", alpha=0.5, label='Radio')
df['Newspaper'].plot.hist(bins=10, color="purple", alpha=0.5, label='Newspaper')
plt.title('Histograms of Advertising Budgets')
plt.xlabel('Budget Allocation')
plt.legend()
plt.show()


corr_matrix = df.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()


from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df[['TV']], df[['Sales']], test_size=0.3, random_state=0)


print("Training data (X):")
print(X_train)
print("Training data (y):")
print(y_train)


from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)


y_pred = model.predict(X_test)
print("Predictions on test data:")
print(y_pred)


coef = model.coef_
intercept = model.intercept_
print("Coefficient:", coef)
print("Intercept:", intercept)


plt.scatter(X_test, y_test, label='Test Data')
plt.plot(X_test, intercept + coef * X_test, 'r', label='Regression Line')
plt.title('Linear Regression')
plt.xlabel('TV Advertising Budget')
plt.ylabel('Sales')
plt.legend()
plt.show()
```
